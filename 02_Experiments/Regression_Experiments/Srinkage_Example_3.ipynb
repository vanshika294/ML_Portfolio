## Business Problem:

You're a data scientist at an e-commerce company.

The marketing team spends money on:

- TV Ads
- YouTube Ads
- Instagram Ads
- Facebook Ads

**1. Goal:** Predict sales and understand the effect of each advertising channel.

**2. Challenge:** Marketing campaigns are often launched together.  
When TV spending increases, YouTube and other channels usually increase too.

**3. Why this is tricky:** All ad channels share a **common driver**: the overall marketing budget.

---
## What is the actual modeling problem?

- You try to fit: Sales=β1​TV+β2​YouTube+β3​Instagram+β4​Facebook

- But the problem is: TV ≈ YouTube ≈ Instagram ≈ Facebook (all correlated)

- So the model cannot answer who deserves credit for sales? Because all channels move together.

---
## Why “when TV increases, YouTube increases too, why are they correlated”?

- It does NOT mean:
    - TV causes YouTube 
    - YouTube causes TV 

- It means: Both are driven by the same decision: “we have more money this month”
