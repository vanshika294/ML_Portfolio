## **Business Scenario: Customer Churn**

- Imagine you're a data scientist at a telecom company.

- Management asks: "Can we identify customers likely to leave next month so we can offer them discounts?"

- This is a binary classification problem.

- Customer Outcome:

```
Stays         0
Leaves	      1
```

- The goal of a linear classifier is to draw a boundary that separates likely churners from non-churners.

### **Data**:
| Feature        | Meaning                |
| -------------- | ---------------------- |
| MonthlyCharges | How much customer pays |
| Tenure         | Months with company    |
| SupportCalls   | Number of complaints   |


### Step 1: create synthetic data

````python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 1000

MonthlyCharges = np.random.normal(70, 20, n)
Tenure = np.random.randint(1, 72, n)

# Generate n customers, where the average customer makes about 2 support calls.
SupportCalls = np.random.poisson(2, n)

'''
Why not use a normal distribution?

Suppose we did: np.random.normal(2, 1, n)

Possible output:2.3, 1.8, -0.7, 4.1

Problems:

Support calls should be whole numbers.
Negative calls don't make sense.

Poisson naturally avoids both issues.
'''
````

### Step 2: Hidden reality

````python
score = (
    0.05 * MonthlyCharges
    - 0.04 * Tenure
    + 0.6 * SupportCalls
)
````

### Step 3: Convert Score to Probability
- We use sigmoid function to convert the score to probability.

````python
probability = 1 / (1 + np.exp(-score))
print(probability)
````

### Step 4: Generate Labels

````python
y = np.random.binomial(
    1,
    probability
)
````

#### **Inspect**

````python
for i in range(10):
    print(
        round(probability[i], 2),
        y[i]
    )
````
