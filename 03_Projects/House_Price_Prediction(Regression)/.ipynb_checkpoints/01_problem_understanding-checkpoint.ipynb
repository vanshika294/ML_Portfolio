# **House Price Prediction Project**

## 1. Objective:
To build a regression model that predicts the sale price of residential houses based on their features.

---
## 2. 🧠 Business Problem Statement:

- A real estate company wants to estimate the selling price of a house before listing it in the market.

- Currently, pricing is based on manual estimation and past experience, which can lead to inaccurate pricing.

- We aim to build a machine learning model that predicts house prices using historical housing data.

---
## 3.🎯 What is being predicted?

Target Variable:
- SalePrice (continuous numerical value)

Type of problem:
- Supervised Learning
- Regression problem

---
## 4. 📊 Why this problem matters ?
Accurate house price prediction helps:
- Sellers price houses competitively
- Buyers avoid overpaying
- Real estate companies make better investment decisions

---
## 5. 🧩 Initial assumptions
Initial assumptions about factors affecting house prices:

- Larger living area increases price
- Higher overall quality increases price
- Better location increases price
- More bathrooms/garage space increases price
- Older houses may have lower prices

---
## 6. 📌 ML problem definition 

This is a supervised learning regression problem where we will:
- Train models on historical data
- Learn relationship between features and price
- Evaluate using error metrics like RMSE and MAE
